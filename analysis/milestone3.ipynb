Dataset

The first step of this project would be importing the data set. I used the following code to import the dataset: 


```python
import pandas as pd
import numpy as np
def load_and_process(directory):
    
    df1 = (pd.read_csv(directory).drop(columns = ["price", "weekly_price"]))
    return df1

df = load_and_process('http://data.insideairbnb.com/china/hk/hong-kong/2020-06-15/data/listings.csv.gz')
```

Analysis
---
Having imported the dataset, I proceeded to analyse it. There were two fields which stood out to me at first. The first field is called "reviews_score_rating". This field denotes the average rating of the Airbnb location. This is an integer field. A higher rating would mean that the Airbnb location provides the best experience for the customer. I would consider this to be the primary field for me to compare between Airbnb locations as it directly conveys information about how customers feel about the Airbnb location. The next field is called "number of reviews". This field denotes the number of reviews for the Airbnb location. This is also an integer field. More reviews would mean that the Airbnb location is the most popular. This field does not inform me about the experience of customers as it merely tells me how popular it is. Perhaps it is in a convenient location or the owner has many contacts. However, I believe that I can combine this field with another field, for example "reviews_score_rating", to provide me with information about how good the experience was in the Airbnb location and how popular the Airbnb location is.

I decided to create a scatter plot to display the number of reviews on the x-axis and the ratings on the y-axis. The reason I chose a scatter plot was because there are two different numerical fields and they are arranged in an unordered manner. If there is a value which has a high number of reviews and a high rating, I can hypothesise that this location could be the best Airbnb location in Hong Kong. Having said this, I used the following code to display the scatter plot:

```python
df1 = df[['review_scores_rating','number_of_reviews']]
ax1 = df1.plot.scatter(y='review_scores_rating',x='number_of_reviews',c='DarkBlue')
ax1.set(xlabel="Number of Reviews (Count)")
ax1.set(ylabel="Reviews Score Rating (Score)")
```

On first look, I noticed an "outlier" which had over 700 reviews and a review of between 90 to 100. On closer look at the dataset, I discovered that this Airbnb location had a score of 96 and a total of 757 reviews. This leads me to believe that this location could be the best Airbnb location in terms of number of reviews and highest score. Having a high number of reviews means that a lot of people have stayed at this location and have had a significant impression. Having a high score means that a lot of people have had a good experience at this location. There is no other Airbnb location which can match the same number of reviews and have such a high score. Therefore, I strongly believe that this is the best Airbnb location based on these two attributes.

Next, I decided to create a strip plot to display the number of people each Airbnb location can accomodate on the x-axis and the ratings on the y-axis. The number of people each Airbnb location can accommodate is denoted by the field known as "accommodates". As number of people is a discrete whole number, the field therefore contains only integers. I decided to create a strip plot as I would like to display the different integer numbers for "accommodates" in relation with the ratings. This would allow me to visualise the review score classified by number of people that can be accommodated. I could have a visualization of which Airbnb location has the highest review score and can accommodate the most people, which could suggest that this location is the best Airbnb location in Hong Kong. Having said this, I used the following code to display the strip plot:

```python
import seaborn as sns
df2 = df[['review_scores_rating','accommodates']]
ax2 = sns.stripplot(data=df2,x='accommodates',y='review_scores_rating')
ax2.set(xlabel="Number of People which can be Accommodated (Count)")
ax2.set(ylabel="Reviews Score Rating (Score)")
```

On first look, I noticed the highest review score of 100 was given for an Airbnb location which can accomodate the highest number of people (16). On closer look at the dataset, I was able to find that there were 3 Airbnb locations which fit this category. Among the 3 locations, there was 1 which had the most number of reviews. This leads me to conclude that this location has the highest score, can accommodate the most people, and is the most popular among the 3 locations which has the same score and can accommodate the same number of people.

Lastly, I decided to create a violin plot to display the number of people each Airbnb location can accomodate on the x-axis and the number of reviews per month on the y-axis. As mentioned previously, the number of people each Airbnb location can accommodate is denoted by the field known as "accommodates". The number of reviews per month is denoted by the field known as "reviews_per_month". I would like to display the different integer numbers for "accommodates" in relation with the number of reviews per month. This would allow me to visualise the number of reviews per month classified by number of people that can be accommodated. I could have a visualization of which Airbnb location has the most number reviews per month and can accommodate the most people, which could suggest that this location is the most popular Airbnb location in Hong Kong. Having said this, I used the following code to display the violin plot:

```python
import seaborn as sns
df3 = df[['reviews_per_month','accommodates']]
ax3 = sns.violinplot(data=df3,x='accommodates',y='reviews_per_month')
ax3.set(xlabel="Number of People which can be Accommodated (Count)")
ax3.set(ylabel="Reviews per Month (Count)")
```

On first look, I noticed that the highest number of reviews per month is more than 12.5 and this location can accommodate 4 people. I also noticed that there was a location which could accommodate 14 people and it has a number of reviews per month of over 10. I realised that the second location might have been the best location in terms of number of reviews per month and number of people it can accomodate. This leads me to believe that this could be the most popular Airbnb location.
